```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

# Display dataset 1

```python
df = pd.read_csv('data/dataset1.csv')

print(df.head())
```

```python
plt.figure(figsize=(10, 6))

for col in df.columns[:3]:
    plt.plot(df[col][:100], label=col)

# Add title and labels
plt.title('Spectrum Sensing Data')
plt.xlabel('Time')  # Adjust according to your data
plt.ylabel('Power Level (dB)')

# Add legend to show which line corresponds to which column
plt.legend()

# Display the plot
plt.show()
```

# Display dataset large dataset from Kaggle

```python
df = pd.read_csv('data/dataset_india.csv')

print(df.head())
```

```python
# Convert timestamp to integer
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
start_date = '2023-05-05'
end_date = '2023-05-07'

reduced_df = df[(df['Timestamp'] >= start_date) & (df['Timestamp'] <= end_date)]
reduced_df = reduced_df[reduced_df['Network Type'] == '5G']
```

```python
plt.figure(figsize=(10, 6))

plt.plot(reduced_df['Timestamp'], reduced_df['Signal Strength (dBm)'], label='Signal Strength (dBm)')

date_format = mdates.DateFormatter('%d-%m %H:%M')
plt.gca().xaxis.set_major_formatter(date_format)

# Add title and labels
plt.title('Spectrum Sensing Data')
plt.xlabel('Time')  # Adjust according to your data
plt.ylabel('Signal strength (dBM)')

# Add legend to show which line corresponds to which column
plt.legend()

# Display the plot
plt.show()
```
